```python
import requests
import os

def evaluate_response(query: str, response: str, api_key: str = None) -> dict:
    """
    Evaluate a query-response pair using the Truesight judge endpoint.
    
    Args:
        query: The user's query/question
        response: The chatbot's response to evaluate
        api_key: Truesight API key (defaults to TRUESIGHT_API_KEY env variable)
    
    Returns:
        dict: The evaluation result from the API
    """
    if api_key is None:
        api_key = "ts_sk_NLL3PO7X1jme7r3rOtTFz6h2IQyj44zDr8KDdDFzy0c" #don't do this! demonstration purposes only!!
    
    api_response = requests.post(
        "https://api.truesight.goodeyelabs.com/api/eval/live_1d81e80eae8f4fe7",
        headers={
            "Authorization": f"Bearer {api_key}",
            "Content-Type": "application/json"
        },
        json={
            "inputs": {
                "query": query,
                "response": response
            }
        }
    )
    
    api_response.raise_for_status()  # Raise an error for bad status codes
    return api_response.json()
```

```python
# Example usage
query_example = "Gluten-light recipe - I'm not celiac just sensitive"

recipe_example = """Here's a delicious Gluten-Light Lemon Herb Quinoa Salad perfect for sensitive stomachs!

Ingredients:
- 1 cup quinoa
- 2 cups water
- 1 cucumber, diced
- Fresh parsley and mint
- Lemon juice and olive oil
- Salt and pepper

Quinoa is naturally gluten-light and easy to digest. 
Rinse quinoa, cook in water for 15 minutes. 
Cool, then toss with vegetables and dressing. A light, refreshing meal!"""

# Test the function
result = evaluate_response(query_example, recipe_example)
print(result)
```
